```python
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
colunas_usadas
```

```python
base = pd.read_csv("house-prices.csv", usecols = colunas_usadas)
```

```python
base.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
X = base.drop('price', axis = 1)
y = base.price
```

```python
X.head()
```

```python
y.head()
```

```python
previsores_colunas = colunas_usadas[1:17]
previsores_colunas
```

```python
import tensorflow as tf
```

```python
colunas = [tf.feature_column.numeric_column(key= c) for c in previsores_colunas]
```

```python
print(colunas[1])
```

```python
from sklearn.model_selection import train_test_split #70% treinamento 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
# test_size quantidade passada para teste
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train, batch_size = 32,
                                                        num_epochs= None, shuffle = True)
```

```python
funcao_teste = tf.estimator.inputs.pandas_input_fn(x = X_test, y = y_test, batch_size = 32,
                                                        num_epochs= 10000, shuffle = False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn = funcao_treinamento, steps = 10000)
```

```python
metricas_treinamento = regressor.evaluate(input_fn = funcao_treinamento, steps=10000)
```

```python
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps = 10000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
funcao_previsao = tf.estimator.inputs.pandas_input_fn(x = X_test, shuffle = False)
```

```python
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
valores_previsoes = []
for p in regressor.predict(input_fn = funcao_previsao):
    valores_previsoes.append(p['predictions'])
```

```python
valores_previsoes
```

```python
import numpy as np
valores_previsoes= np.asarray(valores_previsoes).reshape(-1, 1)
```

```python
valores_previsoes = scaler_y.inverse_transform(valores_previsoes)
valores_previsoes
```

```python
y_test2 = y_test.values.reshape(-1,1)
y_test2.shape
```

```python
y_test2 = scaler_y.inverse_transform(y_test2)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test2, valores_previsoes)
```
